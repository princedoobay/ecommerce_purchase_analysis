**Intermediate Level Analysis**
16. What is the average purchase price for each job category? Are there any specific professions that tend to spend more?

```python
avg_purchase_by_job = df.groupby('Job')['Purchase Price'].mean()
```

```python
print("Average Purchase Price by Job:\n", avg_purchase_by_job)
```

17. Spending behavior based on language


```python
avg_purchase_by_language = df.groupby('Language')['Purchase Price'].mean()
print("\nAverage Purchase Price by Language:\n", avg_purchase_by_language)
```

18. Difference in purchase amounts between AM and PM


```python
avg_purchase_by_time = df.groupby('AM or PM')['Purchase Price'].mean()
print("\nAverage Purchase Price by AM/PM:\n", avg_purchase_by_time)
```

19. Most frequently used credit card providers


```python
most_common_cc_provider = df['CC Provider'].value_counts().head(10)
print("\nMost Common Credit Card Providers:\n", most_common_cc_provider)
```

20. Patterns in purchase prices related to credit card expiration


```python
df['CC Exp Year'] = df['CC Exp Date'].apply(lambda x: int(x.split('/')[1]))
avg_purchase_by_exp_year = df.groupby('CC Exp Year')['Purchase Price'].mean()
print("\nAverage Purchase Price by Credit Card Expiration Year:\n", avg_purchase_by_exp_year)
```

21. Correlation between location (IP Address) and purchase price


```python
avg_purchase_by_ip = df.groupby('IP Address')['Purchase Price'].mean()
print("\nAverage Purchase Price by IP Address:\n", avg_purchase_by_ip.head(10))
```

22. Spending analysis based on email domains

```python
df['Email Domain'] = df['Email'].apply(lambda x: x.split('@')[1])
avg_purchase_by_email_domain = df.groupby('Email Domain')['Purchase Price'].mean().head(10)
print("\nAverage Purchase Price by Email Domain:\n", avg_purchase_by_email_domain)
```

23. Browser-based analysis of purchase amounts


```python
df['Browser'] = df['Browser Info'].apply(lambda x: x.split()[0])  # Extract browser name
avg_purchase_by_browser = df.groupby('Browser')['Purchase Price'].mean()
print("\nAverage Purchase Price by Browser:\n", avg_purchase_by_browser)
```

24. Company-wise purchase price analysis


```python
avg_purchase_by_company = df.groupby('Company')['Purchase Price'].mean().sort_values(ascending=False).head(10)
print("\nTop Companies by Average Purchase Price:\n", avg_purchase_by_company)
```

25. Security risks by company


```python
potential_security_risks = df.groupby('Company')['CC Security Code'].nunique().sort_values(ascending=False).head(10)
print("\nCompanies with Highest Number of Unique Security Codes:\n", potential_security_risks)
```
